```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df= df.drop("Unnamed: 0",axis=1)
print(df.head())
```

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace(" 5 - 10"," 5 - 10")
print(df.head())
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df, x= "Gender")
ax.bar_label(ax.containers[0])
plt.title("GENDER DISTRIBUTION")
plt.show()
```

```python
gp= df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gp)
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(gp,annot=True)
plt.title("RELATION BETWEEN PARENT'S EDUCATION AND STUDENT'S SCORE ")
plt.show()
```

```python
gp1= df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gp1)
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(gp1,annot=True)
plt.title("RELATION BETWEEN PARENT'S  MARITAL STATUS AND STUDENT'S SCORE ")
plt.show()
```

```python
sns.boxplot(data =df ,x ="MathScore")
plt.show()
```
